## LOAD DATASET

```python
import pandas as pd
```

```python
file_url = "https://drive.google.com/uc?export=download&id=1qFeqxx-aYGc4hZ1c59F4dF0IT1XkSCGe"

try:
    df = pd.read_csv(file_url)
    print(f"Successfully reading file from {file_url}")
except Exception as e:
    print(f"Error reading file from {file_url}: {str(e)}")
```

```python
df.info()
df.head()
```

```python
print(f"Total duplicated: ", df.duplicated().sum())
print(f"Total missing values:\n",df.isnull().sum())
```

## EXPLORE DATA

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.info()
```

```python
df.head(10)
```

```python
df.describe()
```

```python
by_order_id = df.groupby("order_id").count()
by_order_id
```

```python
all_numerical = df[['quantity', 'unit_price', 'total_price']]
print(all_numerical)
```

```python
corr_matrix = all_numerical.corr()
```

```python
# Set up the matplotlib figure
plt.figure(figsize=(8, 6))
# Create a heatmap
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={"shrink": .8})
plt.show()
```
